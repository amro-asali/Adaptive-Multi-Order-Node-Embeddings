```python
import torch
import torch.nn.functional as F
from torch_geometric.datasets import Planetoid
from torch_geometric.nn import JumpingKnowledge, GCNConv, MixHopConv
import torch_geometric.transforms as T
import time
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, accuracy_score
import random

def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)

class GCN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, num_layers=6, dropout=0.5):
        super(GCN, self).__init__()
        self.dropout = dropout
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 1):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))

    def forward(self, x, edge_index):
        for i, conv in enumerate(self.convs):
            x = conv(x, edge_index)
            if i < len(self.convs) - 1:
                x = F.relu(x)
                x = F.dropout(x, p=self.dropout, training=self.training)
        return x

class JKNetMax(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels=16, num_layers=6, mode='max', dropout=0.5):
        super(JKNetMax, self).__init__()
        self.dropout = dropout
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 1):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))
        self.jump = JumpingKnowledge(mode=mode, channels=hidden_channels, num_layers=num_layers)

    def forward(self, x, edge_index):
        xs = []
        for conv in self.convs:
            x = conv(x, edge_index)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            xs.append(x)
        return self.jump(xs)

class MixHopEncoder(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels=32, dropout=0.5, powers=[0, 1, 2]):
        super(MixHopEncoder, self).__init__()
        self.conv1 = MixHopConv(in_channels, hidden_channels, powers=powers)

    def forward(self, x, edge_index):
        return self.conv1(x, edge_index)

class AlternatingGCNMixHopJKNet(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, num_layers=2, hop_hidden_channels=32, mode='max', dropout=0.4, mixhop_powers=[0,1,2]):
        super(AlternatingGCNMixHopJKNet, self).__init__()
        self.dropout = dropout
        self.num_layers = num_layers
        self.convs = torch.nn.ModuleList()
        self.lins = torch.nn.ModuleList()
        for i in range(num_layers):
            if i % 2 == 0:
                in_c = in_channels if i == 0 else hidden_channels
                self.convs.append(GCNConv(in_c, hidden_channels))
                self.lins.append(torch.nn.Identity())
            else:
                self.convs.append(MixHopConv(hidden_channels, hop_hidden_channels, powers=mixhop_powers))
                self.lins.append(torch.nn.Linear(hop_hidden_channels * len(mixhop_powers), hidden_channels))
        self.jump = JumpingKnowledge(mode=mode, channels=hidden_channels, num_layers=num_layers)

    def forward(self, x, edge_index):
        xs = []
        for i in range(self.num_layers):
            if i % 2 == 0:
                x = self.convs[i](x, edge_index)
                x = F.relu(x)
            else:
                x = self.convs[i](x, edge_index)
                x = self.lins[i](x)
                x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            xs.append(x)
        return self.jump(xs)


class MultiTaskGNN(torch.nn.Module):
    def __init__(self, encoder, embedding_dim, num_classes):
        super().__init__()
        self.encoder = encoder
        self.classification_head = torch.nn.Linear(embedding_dim, num_classes)
        self.link_prediction_head = torch.nn.Sequential(
            torch.nn.Linear(2 * embedding_dim, 32),
            torch.nn.ReLU(),
            torch.nn.Dropout(p=0.5),
            torch.nn.Linear(32, 1)
        )

    def forward(self, x, edge_index):
        return self.encoder(x, edge_index)

def train_multitask(model, optimizer, original_data, link_train_data):
    model.train()
    optimizer.zero_grad()
    z = model(link_train_data.x, link_train_data.edge_index)
    task = random.randint(0, 1)
    if task == 1:
        node_pred = F.log_softmax(model.classification_head(z), dim=-1)
        loss = F.nll_loss(node_pred[original_data.train_mask], original_data.y[original_data.train_mask])
    else:
        link_pred_input = torch.cat([z[link_train_data.edge_label_index[0]], z[link_train_data.edge_label_index[1]]], dim=-1)
        link_pred = model.link_prediction_head(link_pred_input).squeeze()
        loss = F.binary_cross_entropy_with_logits(link_pred, link_train_data.edge_label)
    loss.backward()
    optimizer.step()
    with torch.no_grad():
        node_pred = F.log_softmax(model.classification_head(z), dim=-1)
        loss_node = F.nll_loss(node_pred[original_data.train_mask], original_data.y[original_data.train_mask]).item()
        link_pred_input = torch.cat([z[link_train_data.edge_label_index[0]], z[link_train_data.edge_label_index[1]]], dim=-1)
        link_pred = model.link_prediction_head(link_pred_input).squeeze()
        loss_link = F.binary_cross_entropy_with_logits(link_pred, link_train_data.edge_label).item()
    return loss_node, loss_link

@torch.no_grad()
def test_multitask(model, original_data, link_train_data, link_val_data, link_test_data):
    model.eval()
    z = model(link_train_data.x, link_train_data.edge_index)
    node_pred = model.classification_head(z).argmax(dim=-1)
    val_acc = accuracy_score(original_data.y[original_data.val_mask].cpu(), node_pred[original_data.val_mask].cpu())
    test_acc = accuracy_score(original_data.y[original_data.test_mask].cpu(), node_pred[original_data.test_mask].cpu())
    val_link_input = torch.cat([z[link_val_data.edge_label_index[0]], z[link_val_data.edge_label_index[1]]], dim=-1)
    val_link_pred = model.link_prediction_head(val_link_input).squeeze().sigmoid()
    val_auc = roc_auc_score(link_val_data.edge_label.cpu(), val_link_pred.cpu())
    test_link_input = torch.cat([z[link_test_data.edge_label_index[0]], z[link_test_data.edge_label_index[1]]], dim=-1)
    test_link_pred = model.link_prediction_head(test_link_input).squeeze().sigmoid()
    test_auc = roc_auc_score(link_test_data.edge_label.cpu(), test_link_pred.cpu())
    return val_acc, test_acc, val_auc, test_auc

def plot_multitask_metrics(history, model_name, seed):
    epochs = range(1, len(history['loss_node']) + 1)
    plt.figure(figsize=(18, 6))
    plt.subplot(1, 3, 1)
    plt.plot(epochs, history['loss_node'], 'y-', label='Node Loss')
    plt.plot(epochs, history['loss_link'], 'c-', label='Link Loss')
    plt.title(f'Losses for {model_name}'); plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.grid(True); plt.legend()
    plt.subplot(1, 3, 2)
    plt.plot(epochs, history['val_acc'], 'r-', label='Val Accuracy')
    plt.plot(epochs, history['test_acc'], 'g-', label='Test Accuracy')
    plt.title(f'Node Classification Accuracy'); plt.xlabel('Epochs'); plt.ylabel('Accuracy'); plt.grid(True); plt.legend()
    plt.subplot(1, 3, 3)
    plt.plot(epochs, history['val_auc'], 'r-', label='Val AUC')
    plt.plot(epochs, history['test_auc'], 'g-', label='Test AUC')
    plt.title(f'Link Prediction AUC'); plt.xlabel('Epochs'); plt.ylabel('AUC'); plt.grid(True); plt.legend()
    plt.suptitle(f'Multi-Task Metrics for {model_name} (Seed: {seed})')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def run(encoder_class_lambda, model_name, seed, embedding_dim):
    set_seed(seed)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    dataset = Amazon(root='./AmazonPhoto', name='Photo', transform=T.NormalizeFeatures())
    data = dataset[0].to(device)
    num_nodes = data.num_nodes
    perm = torch.randperm(num_nodes)
    train_mask = torch.zeros(num_nodes, dtype=torch.bool)
    val_mask = torch.zeros(num_nodes, dtype=torch.bool)
    test_mask = torch.zeros(num_nodes, dtype=torch.bool)
    train_mask[perm[:int(0.6*num_nodes)]] = True
    val_mask[perm[int(0.6*num_nodes):int(0.8*num_nodes)]] = True
    test_mask[perm[int(0.8*num_nodes):]] = True
    
    data.train_mask = train_mask
    data.val_mask = val_mask
    data.test_mask = test_mask
    original_data=data
    transform = T.RandomLinkSplit(num_val=0.05, num_test=0.1, is_undirected=True, add_negative_train_samples=True)
    link_train_data, link_val_data, link_test_data = transform(original_data.clone())
    encoder = encoder_class_lambda(dataset.num_features).to(device)
    model = MultiTaskGNN(encoder, embedding_dim, dataset.num_classes).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)
    best_val_score, final_test_acc, final_test_auc, patience = 0, 0, 0, 10000
    wait = 0
    history = {'loss_node': [], 'loss_link': [], 'val_acc': [], 'test_acc': [], 'val_auc': [], 'test_auc': []}
    for epoch in range(1, 10000):
        try:
            loss_node, loss_link = train_multitask(model, optimizer, original_data.to(device), link_train_data.to(device))
            val_acc, test_acc, val_auc, test_auc = test_multitask(model, original_data.to(device), link_train_data.to(device), link_val_data.to(device), link_test_data.to(device))
            history['loss_node'].append(loss_node)
            history['loss_link'].append(loss_link)
            history['val_acc'].append(val_acc); history['test_acc'].append(test_acc)
            history['val_auc'].append(val_auc); history['test_auc'].append(test_auc)
            current_val_score = val_acc + val_auc
            if current_val_score > best_val_score:
                best_val_score = current_val_score
                final_test_acc = test_acc
                final_test_auc = test_auc
                wait = 0
            else:
                wait += 1
                if wait >= patience:
                    print(f"{model_name} Early stop at epoch {epoch}")
                    break
            if epoch % 100 == 0:
                print(f"Epoch {epoch:03d} | Node Loss: {loss_node:.4f} | Link Loss: {loss_link:.4f} | Val Acc: {val_acc:.4f} | Val AUC: {val_auc:.4f}")
        except RuntimeError as e:
            print(f"Error during training/testing for {model_name} at epoch {epoch}: {e}")
            break
    print(f"\n--- Results for {model_name} (Seed {seed}) ---")
    print(f"Best Combined Val Score (Acc+AUC): {best_val_score:.4f}")
    print(f"Final Test Accuracy @ Best Val = {final_test_acc:.4f}")
    print(f"Final Test AUC @ Best Val      = {final_test_auc:.4f}\n")
    plot_multitask_metrics(history, model_name, seed)

if __name__ == '__main__':
    seeds = [123]
    print("===== Print Num Params =====")
    
    print("===== Regular GCN (Multi-Task) =====")
    run(lambda in_c: GCN(in_c, hidden_channels=32, num_layers=2, dropout=0.5), "GCN", seeds[0], embedding_dim=32)
    
    print("\n===== JKNet Max Pool (Multi-Task) =====")
    run(lambda in_c: JKNetMax(in_c, hidden_channels=32, num_layers=2, dropout=0.5), "JKNetMax", seeds[0], embedding_dim=32)

    print("\n===== JKNet lstm (Multi-Task) =====")
    run(lambda in_c: JKNetMax(in_c, hidden_channels=32, num_layers=2, mode='lstm', dropout=0.5), "JKNetMax", seeds[0], embedding_dim=32)
    
    print("\n===== MixHop (Multi-Task) =====")
    run(lambda in_c: MixHopEncoder(in_c, hidden_channels=32, dropout=0.3), "MixHop", seeds[0], embedding_dim=32*3)
    
    print("\n===== AlternatingGCNMixHopJKNet (Multi-Task) =====")
    run(lambda in_c: AlternatingGCNMixHopJKNet(in_c, hidden_channels=32, num_layers=2, hop_hidden_channels=32,\
    mode='max', dropout=0.5), "AltGCNMixHopJKN", seeds[0], embedding_dim=32)
```

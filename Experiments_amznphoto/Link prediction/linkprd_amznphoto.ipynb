```python

"""LINK PREDICTION"""
import torch
import torch.nn.functional as F
from torch_geometric.datasets import Planetoid
from torch_geometric.nn import JumpingKnowledge, GCNConv, MixHopConv
import torch_geometric.transforms as T
import time
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score

def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)


class GCN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, num_layers=6, dropout=0.5):
        super(GCN, self).__init__()
        self.dropout = dropout
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 1):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))

    def forward(self, x, edge_index):
        for i, conv in enumerate(self.convs):
            x = conv(x, edge_index)
            if i < len(self.convs) - 1:
                x = F.relu(x)
                x = F.dropout(x, p=self.dropout, training=self.training)
        return x

class JKNetMax(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels=16, num_layers=6, dropout=0.5):
        super(JKNetMax, self).__init__()
        self.dropout = dropout
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 1):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))
        self.jump = JumpingKnowledge(mode='max', channels=hidden_channels, num_layers=num_layers)

    def forward(self, x, edge_index):
        xs = []
        for conv in self.convs:
            x = conv(x, edge_index)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            xs.append(x)
        return self.jump(xs)

class MixHopEncoder(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels=32, dropout=0.5, powers=[0, 1, 2]):
        super(MixHopEncoder, self).__init__()
        self.conv1 = MixHopConv(in_channels, hidden_channels, powers=powers)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        return x


class AlternatingGCNMixHopJKNet(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, num_layers=2, hop_hidden_channels=32, mode='max', dropout=0.4, mixhop_powers=[0,1,2]):
        super(AlternatingGCNMixHopJKNet, self).__init__()
        self.dropout = dropout
        self.num_layers = num_layers
        self.convs = torch.nn.ModuleList()
        self.lins = torch.nn.ModuleList()
        for i in range(num_layers):
            if i % 2 == 0:
                in_c = in_channels if i == 0 else hidden_channels
                self.convs.append(GCNConv(in_c, hidden_channels))
                self.lins.append(torch.nn.Identity())
            else:
                self.convs.append(MixHopConv(hidden_channels, hop_hidden_channels, powers=mixhop_powers))
                self.lins.append(torch.nn.Linear(hop_hidden_channels * len(mixhop_powers), hidden_channels))
        self.jump = JumpingKnowledge(mode=mode, channels=hidden_channels, num_layers=num_layers)

    def forward(self, x, edge_index):
        xs = []
        for i in range(self.num_layers):
            if i % 2 == 0:
                x = self.convs[i](x, edge_index)
                x = F.relu(x)
            else:
                x = self.convs[i](x, edge_index)
                x = self.lins[i](x)
                x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            xs.append(x)
        return self.jump(xs)

class LinkPredictor(torch.nn.Module):
    def __init__(self, encoder, decoder_in_channels, decoder_hidden_channels=128):
        super().__init__()
        self.encoder = encoder
        self.decoder = torch.nn.Sequential(
            torch.nn.Linear(decoder_in_channels, decoder_hidden_channels),
            torch.nn.ReLU(),
            torch.nn.Dropout(p=0.5),
            torch.nn.Linear(decoder_hidden_channels, 1)
        )

    def forward(self, x, edge_index, edge_label_index):
        z = self.encoder(x, edge_index)
        
        src_emb = z[edge_label_index[0]]
        dst_emb = z[edge_label_index[1]]
        emb_cat = torch.cat([src_emb, dst_emb], dim=-1)
        
        return self.decoder(emb_cat).squeeze()

def train(model, optimizer, data):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index, data.edge_label_index)
    loss = F.binary_cross_entropy_with_logits(out, data.edge_label)
    loss.backward()
    optimizer.step()
    return float(loss)

@torch.no_grad()
def test(model, train_data, data_to_test):
    model.eval()
    with torch.no_grad():
        out = model(train_data.x, train_data.edge_index, data_to_test.edge_label_index)
    out = out.sigmoid()
    return roc_auc_score(data_to_test.edge_label.cpu().numpy(), out.cpu().numpy())

def plot_metrics(history, model_name, seed):
    epochs = range(1, len(history['loss']) + 1)
    plt.figure(figsize=(15, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['loss'], 'y-', label='Training Loss')
    plt.title(f'Loss for {model_name}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['val_auc'], 'r-', label='Validation AUC')
    plt.plot(epochs, history['test_auc'], 'g-', label='Test AUC')
    plt.title(f'AUC for {model_name}')
    plt.xlabel('Epochs')
    plt.ylabel('AUC')
    plt.grid(True)
    plt.legend()
    plt.suptitle(f'Link Prediction Metrics for {model_name} (Seed: {seed})')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

def run(encoder_class_lambda, model_name, seed, embedding_dim):
    set_seed(seed)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    dataset = Amazon(root='./AmazonPhoto', name='Photo', transform=T.NormalizeFeatures())
    data = dataset[0].to(device)
    num_nodes = data.num_nodes
    perm = torch.randperm(num_nodes)
    train_mask = torch.zeros(num_nodes, dtype=torch.bool)
    val_mask = torch.zeros(num_nodes, dtype=torch.bool)
    test_mask = torch.zeros(num_nodes, dtype=torch.bool)
    train_mask[perm[:int(0.6*num_nodes)]] = True
    val_mask[perm[int(0.6*num_nodes):int(0.8*num_nodes)]] = True
    test_mask[perm[int(0.8*num_nodes):]] = True
    
    data.train_mask = train_mask
    data.val_mask = val_mask
    data.test_mask = test_mask
    
    transform = T.RandomLinkSplit(num_val=0.05, num_test=0.1, is_undirected=True, add_negative_train_samples=True)
    train_data, val_data, test_data = transform(data)
    
    encoder = encoder_class_lambda(dataset.num_features).to(device)
    
    model = LinkPredictor(encoder, 2 * embedding_dim).to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=5e-4)

    best_val_auc, final_test_auc, patience = 0, 0, 10000
    wait = 0
    history = {'loss': [], 'val_auc': [], 'test_auc': []}

    for epoch in range(1, 10000):
        loss = train(model, optimizer, train_data.to(device))
        val_auc = test(model, train_data.to(device), val_data.to(device))
        test_auc = test(model, train_data.to(device), test_data.to(device))
        
        history['loss'].append(loss)
        history['val_auc'].append(val_auc)
        history['test_auc'].append(test_auc)

        if val_auc > best_val_auc:
            best_val_auc = val_auc
            final_test_auc = test_auc
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                print(f"{model_name} Early stop at epoch {epoch}")
                break
        
        if epoch % 100 == 0:
            print(f"{model_name} Epoch {epoch:03d}, Loss {loss:.4f}, Val AUC {val_auc:.4f}, Test AUC {test_auc:.4f}")

    print(f"{model_name} Seed {seed} â†’ Best Val AUC {best_val_auc:.4f}, Test @BestVal = {final_test_auc:.4f}\n")
    plot_metrics(history, model_name, seed)
    return final_test_auc

if __name__ == '__main__':
    seeds = [123]
    
    dataset = Planetoid(root='./Cora', name='Cora', transform=T.NormalizeFeatures())
    in_c = dataset.num_features

    models = [
        ('GCN', GCN(in_c, hidden_channels=32, num_layers=6, dropout=0.5)),
        ('JKNet', JKNetMax(in_c, hidden_channels=32, num_layers=6, dropout=0.5)),
        ('MixHop', MixHopEncoder(in_c, hidden_channels=32, dropout=0.5)),
        ('AltGCNMixJKN', AlternatingGCNMixHopJKNet(in_c, hidden_channels=32, num_layers=2, hop_hidden_channels=32, mode='max', dropout=0.4))
    ]
    
    for name, model in models:
        num_params = sum(p.numel() for p in model.parameters())
        print(f'{name}: {num_params:,} parameters')
    
    telescopic_results = []
    
    print("===== Regular 2-layer GCN =====")
    gcn_results = [run(lambda in_c: GCN(in_c, hidden_channels=32, num_layers=2, dropout=0.5), "GCN", s, embedding_dim=32) for s in seeds]
    
    print("===== JKNet Max Pool =====")
    jknet_results = [run(lambda in_c: JKNetMax(in_c, hidden_channels=32, num_layers=6, dropout=0.5), "JKNetMax", s, embedding_dim=32) for s in seeds]
    
    print("===== MixHop =====")
    mixhop_results = [run(lambda in_c: MixHopEncoder(in_c, hidden_channels=32, dropout=0.5), "MixHop", s, embedding_dim=96) for s in seeds]

    print("===== alternating MixHop JKNet =====")
    alternating_results = [run(lambda in_c: AlternatingGCNMixHopJKNet(in_c, hidden_channels=32, num_layers=3, hop_hidden_channels=32, mode='max', dropout=0.4), "AlternatingGCNMixHopJKNet", s, embedding_dim=32) for s in seeds]

    print(f"GCN Mean Test AUC = {sum(gcn_results)/len(gcn_results):.4f} Â± {torch.std(torch.tensor(gcn_results), unbiased=False):.4f}")
    print(f"JKNetMax Mean Test AUC = {sum(jknet_results)/len(jknet_results):.4f} Â± {torch.std(torch.tensor(jknet_results), unbiased=False):.4f}")
    print(f"MixHop Mean Test AUC = {sum(mixhop_results)/len(mixhop_results):.4f} Â± {torch.std(torch.tensor(mixhop_results), unbiased=False):.4f}")
    if telescopic_results:
        print(f"TelescopicMixHopJKNet Mean Test AUC = {sum(telescopic_results)/len(telescopic_results):.4f} Â± {torch.std(torch.tensor(telescopic_results), unbiased=False):.4f}")
    print(f"AlternatingMixHopJKNet Mean Test AUC = {sum(alternating_results)/len(alternating_results):.4f} Â± {torch.std(torch.tensor(alternating_results), unbiased=False):.4f}")
```

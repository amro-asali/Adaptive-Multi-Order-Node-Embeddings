```python

"""NODE CLASSIFICATION"""

import torch
import torch.nn.functional as F
from torch_geometric.datasets import Planetoid
from torch_geometric.nn import JumpingKnowledge, GCNConv, MixHopConv
import torch_geometric.transforms as T
import time
import matplotlib.pyplot as plt

def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)

# --- 1. GCN ---
class GCN(torch.nn.Module):
    def __init__(self, in_channels, out_channels, hidden_channels=32, num_layers=2, dropout=0.5):
        super(GCN, self).__init__()
        self.dropout = dropout
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 2):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))
        self.convs.append(GCNConv(hidden_channels, out_channels))

    def forward(self, x, edge_index):
        for conv in self.convs[:-1]:
            x = conv(x, edge_index)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.convs[-1](x, edge_index)
        return F.log_softmax(x, dim=-1)

# --- 2. JKNet ---
class JKNetMax(torch.nn.Module):
    def __init__(self, in_channels, out_channels, hidden_channels=32, num_layers=6, dropout=0.5):
        super(JKNetMax, self).__init__()
        self.dropout = dropout
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 1):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))
        self.jump = JumpingKnowledge(mode='max', channels=hidden_channels, num_layers=num_layers)
        self.classifier = torch.nn.Linear(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        xs = []
        for conv in self.convs:
            x = conv(x, edge_index)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            xs.append(x)
        x_jump = self.jump(xs)
        out = self.classifier(x_jump)
        return F.log_softmax(out, dim=-1)

class MixHopEncoder(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, dropout=0.5, powers=[0, 1, 2]):
        super(MixHopEncoder, self).__init__()
        self.dropout = dropout
        self.conv1 = MixHopConv(in_channels, hidden_channels, powers=powers)
        self.final_lin = torch.nn.Linear(hidden_channels * len(powers), out_channels)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.final_lin(x)
        return F.log_softmax(x, dim=-1)


class AlternatingGCNMixHopJKNet(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, num_layers=4, hop_hidden_channels=16, mode='max', dropout=0.5, mixhop_powers=[0,1,2]):
        super(AlternatingGCNMixHopJKNet, self).__init__()
        self.dropout = dropout
        self.num_layers = num_layers
        self.convs = torch.nn.ModuleList()
        self.lins = torch.nn.ModuleList()
        for i in range(num_layers):
            if i % 2 == 0:
                # GCN layer
                in_c = in_channels if i == 0 else hidden_channels
                self.convs.append(GCNConv(in_c, hidden_channels))
                self.lins.append(torch.nn.Identity())
            else:
                # MixHop layer
                self.convs.append(MixHopConv(hidden_channels, hop_hidden_channels, powers=mixhop_powers))
                self.lins.append(torch.nn.Linear(hop_hidden_channels * len(mixhop_powers), hidden_channels))
        self.jump = JumpingKnowledge(mode=mode, channels=hidden_channels, num_layers=num_layers)
        if mode == 'cat':
            self.final_layer = torch.nn.Linear(num_layers * hidden_channels, out_channels)
        else:
            self.final_layer = torch.nn.Linear(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        xs = []
        for i in range(self.num_layers):
            if i % 2 == 0:
                x = self.convs[i](x, edge_index)
                x = F.relu(x)
            else:
                x = self.convs[i](x, edge_index)
                x = self.lins[i](x)
                x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            xs.append(x)
        x_jump = self.jump(xs)
        x_final = self.final_layer(x_jump)
        return F.log_softmax(x_final, dim=-1)

# --- 3. Training and Testing Functions ---
def train(model, optimizer, data):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return float(loss)

@torch.no_grad()
def test(model, data):
    model.eval()
    pred = model(data.x, data.edge_index).argmax(dim=-1)
    accs = [(pred[mask] == data.y[mask]).sum().item() / mask.sum().item() 
            for mask in [data.train_mask, data.val_mask, data.test_mask]]
    return accs

# --- Plotting Function ---
def plot_metrics(history, model_name, seed):
    """Plots the training and validation loss and accuracy."""
    epochs = range(1, len(history['train_loss']) + 1)
    
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['train_loss'], 'y-', label='Training Loss')
    plt.plot(epochs, history['val_loss'], 'r-', label='Validation Loss')
    plt.title(f'Loss for {model_name}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['train_acc'], 'y-', label='Training Accuracy')
    plt.plot(epochs, history['val_acc'], 'r-', label='Validation Accuracy')
    plt.title(f'Accuracy for {model_name}')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.legend()

    plt.suptitle(f'Training and Validation Metrics for {model_name} (Seed: {seed})')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


# --- 4. Experiment Function ---
def run(model_class, model_name, seed):
    set_seed(seed)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    dataset = Planetoid(root='./Cora', name='Cora', transform=T.NormalizeFeatures())
    data = dataset[0].to(device)
    if model_name in ["MixHop", "AlternatingMixHopJKNet"]:
        model = model_class(dataset.num_features, dataset.num_classes).to(device)
    else:
        model = model_class(dataset.num_features, dataset.num_classes, 64, 6, 0.5).to(device)
    
    optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=5e-4)

    best_val, final_test, patience = 0, 0, 10000
    wait = 0

    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc': [], 'val_acc': []
    }

    for epoch in range(1, 10000):
        loss = train(model, optimizer, data)
        train_acc, val_acc, test_acc = test(model, data)

        model.eval()
        with torch.no_grad():
            out = model(data.x, data.edge_index)
            val_loss = F.nll_loss(out[data.val_mask], data.y[data.val_mask])

        history['train_loss'].append(loss)
        history['val_loss'].append(val_loss.item())
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        if val_acc > best_val:
            best_val, final_test = val_acc, test_acc
            best_state = model.state_dict()
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                print(f"{model_name} Early stop at epoch {epoch}")
                break

        if epoch % 20 == 0:
            print(f"{model_name} Epoch {epoch:03d}, Loss {loss:.4f}, Val {val_acc:.4f}, Test {test_acc:.4f}")

    model.load_state_dict(best_state)
    print(f"{model_name} Seed {seed} â†’ Best Val {best_val:.4f}, Test @BestVal = {final_test:.4f}\n")
    
    plot_metrics(history, model_name, seed)
    
    return final_test

# --- 5. Run and Compare ---
if __name__ == '__main__':
    seeds = [123]
    telescopic_results = []

    print("===== Regular 2-layer GCN =====")
    gcn_results = [run(GCN, "GCN", s) for s in seeds]
    
    print("===== JKNet Max Pool =====")
    jknet_results = [run(JKNetMax, "JKNetMax", s) for s in seeds]
    
    print("===== MixHop =====")
    mixhop_results = [run(lambda in_c, out_c: MixHopEncoder(in_c, 32, out_c, 0.5), "MixHop", s) for s in seeds]
    
    print("===== alternating MixHop JKNet =====")
    alternating_results = [run(lambda in_c, out_c: AlternatingGCNMixHopJKNet(in_c, 32, out_c, num_layers=3, hop_hidden_channels=32, mode='max', dropout=0.4), "AlternatingMixHopJKNet", s) for s in seeds]

    print(f"GCN Mean Test Accuracy = {sum(gcn_results)/len(gcn_results):.4f} Â± {torch.std(torch.tensor(gcn_results), unbiased=False):.4f}")
    print(f"JKNetMax Mean Test Accuracy = {sum(jknet_results)/len(jknet_results):.4f} Â± {torch.std(torch.tensor(jknet_results), unbiased=False):.4f}")
    print(f"MixHop Mean Test Accuracy = {sum(mixhop_results)/len(mixhop_results):.4f} Â± {torch.std(torch.tensor(mixhop_results), unbiased=False):.4f}")
    if telescopic_results:
        print(f"TelescopicMixHopJKNet Mean Test Accuracy = {sum(telescopic_results)/len(telescopic_results):.4f} Â± {torch.std(torch.tensor(telescopic_results), unbiased=False):.4f}")
    print(f"AlternatingMixHopJKNet Mean Test Accuracy = {sum(alternating_results)/len(alternating_results):.4f} Â± {torch.std(torch.tensor(alternating_results), unbiased=False):.4f}")
```
